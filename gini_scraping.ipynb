```python
browser.get('https://www.genie.co.kr/chart/top200')
```

```python
html = browser.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
tags = soup.select('td.info')
tags
```

```python
len(tags), type(tags)
```

```python
tag = tags[0]
```

### 곡명은 a.title /// 가수이름 a.artist ///

```python
title_text = title[0].text.strip()
```

```python
type(artist), len(artist)
```

```python
artist_text = artist[0].text.strip()
```

```python
contents = []
```

```python
for tag in tags:
    title = tag.select('a.title')
    artist = tag.select('a.artist')
    print(f"{artist[0].text.strip()} - {title[0].text.strip()}")
    song = [artist[0].text.strip() , title[0].text.strip()]
    contents.append(song)
```

```python
contents
```

```python
import pandas as pd
```

```python
pd_data =pd.DataFrame(contents, columns=['artist','title'])
```
